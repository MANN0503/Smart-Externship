```python
import matplotlib.pyplot as plt
```

## **LINE PLOT**

```python
x = [6,2,3,4,4]
y = [1,2,3,4,5]
#line plot
plt.plot(x,y)
```

## **BAR PLOT = display categorical data**

```python
x = [6,2,3,4,4]
y = [1,2,3,4,5]
#bar plot
plt.bar(x,y)
```

## **HISTOGRAM = NO. OF OBSERVATIONS IN EACH INTERVAL**

```python
#histogram
y = [9,5,7,7,7,7,4,4,4,2,5]
plt.hist(y)
```

## **SCATTER PLOT = RELATIONSHIP BETWEEN TWO NUMERICAL VARIABLES**

```python
x = [6,2,3,4,4]
y = [1,2,3,4,5]
#scatter plot
plt.scatter(x,y)
```

```python
x = [1,2,3,4,5]
y = [4,5,6,7,8]
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel('X values')
plt.ylabel('Y values')
```
